# Introducción

En una red lineal de admitancias (o impedancias), es posible expresar cualquier relación de transferencia entre dos nodos en función de la razón de admitancias equivalentes.  
Esto se debe a que, sin importar la cantidad de elementos o ramas presentes, el circuito puede simplificarse a dos admitancias equivalentes ($ Y_1 $ e $Y_2 $) conectadas al nodo de interés.  

De este modo, la **función de transferencia** entre los nodos $ V_1 $ y $ V_2 $ puede escribirse en la forma general:

\begin{equation}
\frac{V_2}{V_1} = \frac{Y_1}{Y_1 + Y_2}
\end{equation}

donde $ Y_1 $ representa la admitancia que conecta el nodo de salida $ V_2 $ con el nodo de entrada $ V_1 $ e $ Y_2 $ la admitancia a tierra.  

Esta relación se presenta por realizar un divisor de tensión (con admitancias) y permite analizar el comportamiento de cualquier red sin importar la topología.  

Además, teniendo en cuenta la simplificación del circuito en dos admitancias, también puede aplicarse el principio de **partición de admitancias**, que establece que un elemento puede representarse fraccionado entre dos nodos distintos si se mantiene la proporción de su efecto eléctrico (es decir, se la puede trasladar a otro potencial).  
Por ejemplo, una capacitancia $ C $ puede dividirse en dos partes $C \cdot a $ y $ C \cdot (1 - a) $, aplicando la fracción $ a $ a un potencial y $ (1 - a) $ al otro.  
Del mismo modo, una inductancia o resistencia pueden representarse mediante sus valores equivalentes $ L/b$ con $ L/(1-b) $ y $ R/d$ con $ R/(1-d) $ respectivamente, como en el circuito de la siguiente figura:

![Circuito](Circuito-TS6.jpg)

A partir de esta simplificación, la transferencia $ V_2/V_1 $ se expresa como una función racional en la variable compleja $ s $, de la forma:

\begin{equation}
T(s) = \frac{P(s)}{Q(s)} = \frac{b_2 s^2 + b_1 s + b_0}{a_2 s^2 + a_1 s + a_0}
\end{equation}

donde $ P(s) $ y $ Q(s) $ son los polinomios de segundo orden asociados al numerador (ceros) y al denominador (polos) respectivamente. 

Los ceros ($ Z_i $) son las raíces de $ P(s) $ y los polos ($ P_i $) las raíces de $ Q(s) $. Sus posiciones en el plano complejo determinan las propiedades dinámicas del sistema, tales como la **frecuencia característica**, el **factor de calidad** y la **respuesta en módulo y fase**.

La parametrización de estos polinomios mediante los parámetros *$\omega_0 $*, *Q* y *k* permite representar la función de transferencia en una forma normalizada, facilitando la construcción de los **diagramas de polos y ceros**, con la siguiente fórmula:

\begin{equation}
T(s) = k \cdot \frac{s^2 + \frac{s \omega_{0Z}}{Q_Z} + \omega_{0Z}^2}{s^2 + \frac{s \omega_{0P}}{Q_P} + \omega_{0P}^2}
\end{equation}

Esta expresión corresponde a una **sección bicuadrática**, es decir, una función de transferencia de segundo orden, que representa el comportamiento más general posible de una red lineal con dos polos y dos ceros.  
A partir de esta forma normalizada pueden derivarse todas las demás secciones de menor o mayor orden (para esta última se utiliza varias secciones en cascada, con sus debidas implicancias).  

Por ejemplo, si alguno de los coeficientes cuadráticos se anula ($ s^2 $), la función se reduce a una sección de primer orden, típica de filtros *RC* o *RL*. Por otro lado, cuando el numerador y el denominador comparten polos o ceros, el sistema se simplifica manteniendo la misma estructura algebraica.  

Por esta razón, el análisis de una red, mediante su sección bicuadrática, permite describir todos los posibles comportamientos dinámicos (ya sean pasa bajos, pasa altos, pasa banda o elimina banda) a partir de los valores de los parámetros *$ \omega_0 $*, *Q* y *k*.



# Análisis y desarrollo

## Resolución del circuito con admitancias

La transferencia está dada por:

\begin{equation}
T(s) = \frac{V_2(s)}{V_1(s)} = \frac{Y_1}{Y_1 + Y_2}
\end{equation}

donde:

- $ Y_1 $ corresponde a la parte fraccionaria $a, b, d$.
- $Y_2$ corresponde a la parte $ (1-a), (1-b), (1-d)$.

---

### Admitancia de cada rama

**Resistencia:**

\begin{equation}
R/d \;\Rightarrow\; Y_{R1} = \frac{d}{R}
\end{equation}

**Inductor:**

\begin{equation}
L/b \;\Rightarrow\; Z_{L1} = s \frac{L}{b} \;\Rightarrow\; Y_{L1} = \frac{b}{sL}
\end{equation}

**Capacitor:**

\begin{equation}
C a \;\Rightarrow\; Y_{C1} = s a C
\end{equation}

Por lo tanto:

\begin{equation}
Y_1 = \frac{d}{R} + s a C + \frac{b}{sL}
\end{equation}

La otra rama con los complementos $(1-d, 1-b, 1-a)$ es:

\begin{equation}
Y_2 = \frac{1-d}{R} + s(1-a)C + \frac{1-b}{sL}
\end{equation}

---

La transferencia entre los nodos señalados (divisor de admitancias) es:

\begin{equation}
T(s) = \frac{Y_1}{Y_1 + Y_2}
\end{equation}

Se observa que:

\begin{equation}
Y_1 + Y_2 = \frac{d + (1-d)}{R} + sC(a + (1-a)) + \frac{b + (1-b)}{sL}
\end{equation}

por lo que resulta independiente de $ a, b, d $:

\begin{equation}
Y_1 + Y_2 = \frac{1}{R} + sC + \frac{1}{sL}
\end{equation}

---

### Transformación a polinomios racionales

Multiplicamos numerador y denominador por $sL$ para eliminar los términos $1/(sL)$:

**Numerador:**

\begin{equation}
sL \, Y_1 = sL \left( \frac{d}{R} + s a C + \frac{b}{sL} \right)
\end{equation}

\begin{equation}
sL \, Y_1 = s^2 (L a C) + s \left( \frac{L d}{R} \right) + b
\end{equation}

**Denominador:**

\begin{equation}
sL (Y_1 + Y_2) = sL \left( \frac{1}{R} + sC + \frac{1}{sL} \right)
\end{equation}

\begin{equation}
sL (Y_1 + Y_2) = s^2 (L C) + s \left( \frac{L}{R} \right) + 1
\end{equation}

Por lo tanto:

\begin{equation}
T(s) = \frac{s^2 (L a C) + s \left( \frac{L d}{R} \right) + b}{s^2 (L C) + s \left( \frac{L}{R} \right) + 1}
\end{equation}

---

### Conversión a polinomios mónicos

Dividimos numerador y denominador por $LC$ para tener coeficiente 1 delante de $s^2$ en el denominador:

**Denominador:**

\begin{equation}
\frac{1}{L C} \left( s^2 (L C) + s \left( \frac{L}{R} \right) + 1 \right)
\end{equation}

\begin{equation}
= s^2 + s \left( \frac{1}{R C} \right) + \frac{1}{L C}
\end{equation}

Por lo tanto:

\begin{equation}
\omega_{0P}^2 = \frac{1}{L C}, \qquad \frac{\omega_{0P}}{Q_P} = \frac{1}{R C}
\end{equation}

De esta manera, se obtiene:

\begin{equation}
\omega_{0P} = \frac{1}{\sqrt{L C}}, \qquad Q_P = \frac{R}{L} \sqrt{L C}
\end{equation}

**Numerador (también dividido por $LC$):**

\begin{equation}
\frac{1}{L C} \left( s^2 (L a C) + s \left( \frac{L d}{R} \right) + b \right)
\end{equation}

\begin{equation}
= a s^2 + s \left( \frac{d}{R C} \right) + \frac{b}{L C}
\end{equation}

---

### Forma normalizada de segundo orden

Escribimos el numerador en forma estándar:

\begin{equation}
T(s) = k \cdot \frac{s^2 + \frac{s \omega_{0Z}}{Q_Z} + \omega_{0Z}^2}{s^2 + \frac{s \omega_{0P}}{Q_P} + \omega_{0P}^2}
\end{equation}

Identificando coeficientes, el término principal da:

\begin{equation}
k = a
\end{equation}

Comparando el resto (habiendo sacado factor común *a*):

\begin{equation}
\omega_{0Z}^2 = \frac{b}{a L C}, \qquad \frac{\omega_{0Z}}{Q_Z} = \frac{d}{a R C}
\end{equation}

Por lo tanto:

\begin{equation}
\omega_{0Z} = \sqrt{\frac{b}{a L C}}
\end{equation}

\begin{equation}
Q_Z = \frac{a R C \, \omega_{0Z}}{d} = \frac{R}{L} \sqrt{a b C L} \cdot \frac{1}{d}
\end{equation}

---

### Relación entre ceros y polos

Como:

\begin{equation}
\omega_{0P}^2 = \frac{1}{L C}
\end{equation}

tenemos:

\begin{equation}
\omega_{0Z} = \sqrt{\frac{b}{a}} \, \omega_{0P}
\end{equation}

y usando $Q_P = \frac{R}{L} \sqrt{L C}$, se cumple:

\begin{equation}
Q_Z = Q_P \cdot \frac{ \sqrt{a b}}{d}
\end{equation}

---

### Resultados finales

\begin{equation}
T(s) = k \cdot
\frac{s^2 + \dfrac{s}{Q_Z \omega_{0Z}} + \omega_{0Z}^2}
{s^2 + \dfrac{s}{Q_P \omega_{0P}} + \omega_{0P}^2}
\end{equation}

con:

\begin{equation}
\boxed{
\begin{aligned}
k &= a \\
\omega_{0P} &= \frac{1}{\sqrt{L C}} \\
Q_P &= \frac{R}{L} \sqrt{L C} \\
\omega_{0Z} &= \sqrt{\frac{b}{a}} \, \omega_{0P} \\
Q_Z &= Q_P \, \frac{ \sqrt{a b}}{d}
\end{aligned}
}
\end{equation}

### Diagrama de Polos y Ceros, Gráficos de la Respuesta de Módulo y Fase

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Función para formatear el eje Y en múltiplos de π
def format_func(value, tick_number):
    return f'{value:.1f}π'

#%% T1
num1 = [1, 0, 9]
den1 = [1, np.sqrt(2), 1]

# Respuesta en frecuencia
w1, h1 = signal.freqs(num1, den1, worN=np.logspace(-1, 2, 1000))

# Cálculo de fase
phase1 = np.unwrap(np.angle(h1))

# Polos y ceros
z1, p1, k1 = signal.tf2zpk(num1, den1)

# Crear figura
plt.figure(figsize=(14, 5))

# Magnitud
plt.subplot(1, 3, 1)
plt.semilogx(w1, 20*np.log10(abs(h1)), linewidth=2)
plt.title('Respuesta en Magnitud', fontsize=12, fontweight='bold')
plt.xlabel('Frecuencia angular ω [rad/s]', fontsize=11)
plt.ylabel('|H(jω)| [dB]', fontsize=11)
plt.grid(True, which='both', ls=':', alpha=0.7)

# Fase
plt.subplot(1, 3, 2)
plt.plot(w1, phase1/np.pi, linewidth=2)
plt.title('Respuesta en Fase', fontsize=12, fontweight='bold')
plt.xlabel('Frecuencia angular ω [rad/s]', fontsize=11)
plt.ylabel('∠H(jω) [rad]', fontsize=11)
plt.xlim([0, 10])
plt.grid(True, which='both', ls=':', alpha=0.7)
# Configurar formato de ticks en múltiplos de π
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_func))

# Diagrama de polos y ceros
plt.subplot(1, 3, 3)
plt.plot(np.real(p1), np.imag(p1), 'x', markersize=12, markeredgewidth=2.5,
         label='Polos', color='blue')
if len(z1) > 0:
    plt.plot(np.real(z1), np.imag(z1), 'o', markersize=12, markeredgewidth=2.5,
             fillstyle='none', label='Ceros', color='red')

plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
circle1 = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--', linewidth=1, alpha=0.5)
plt.gca().add_patch(circle1)
plt.title('Diagrama de Polos y Ceros', fontsize=12, fontweight='bold')
plt.xlabel('Parte Real σ [1/s]', fontsize=11)
plt.ylabel('Parte Imaginaria jω [rad/s]', fontsize=11)
plt.grid(True, alpha=0.7)
plt.legend(fontsize=10, loc='best')
plt.axis('equal')

plt.suptitle('T₁(s)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

#%% T2

num2 = [1, 0, 1/9]
den2 = [1, 1/5, 1]

# Respuesta en frecuencia
w2, h2 = signal.freqs(num2, den2, worN=np.logspace(-1, 2, 1000))

# Cálculo de fase
phase2 = np.unwrap(np.angle(h2))

# Polos y ceros
z2, p2, k2 = signal.tf2zpk(num2, den2)

# Crear figura
plt.figure(figsize=(14, 5))

# Magnitud
plt.subplot(1, 3, 1)
plt.semilogx(w2, 20*np.log10(abs(h2)), linewidth=2)
plt.title('Respuesta en Magnitud', fontsize=12, fontweight='bold')
plt.xlabel('Frecuencia angular ω [rad/s]', fontsize=11)
plt.ylabel('|H(jω)| [dB]', fontsize=11)
plt.grid(True, which='both', ls=':', alpha=0.7)

# Fase
plt.subplot(1, 3, 2)
plt.plot(w2, phase2/np.pi, linewidth=2)
plt.title('Respuesta en Fase', fontsize=12, fontweight='bold')
plt.xlabel('Frecuencia angular ω [rad/s]', fontsize=11)
plt.ylabel('∠H(jω) [rad]', fontsize=11)
plt.xlim([0, 10])
plt.grid(True, which='both', ls=':', alpha=0.7)
# Configurar formato de ticks en múltiplos de π
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_func))

# Diagrama de polos y ceros
plt.subplot(1, 3, 3)
plt.plot(np.real(p2), np.imag(p2), 'x', markersize=12, markeredgewidth=2.5,
         label='Polos', color='blue')
if len(z2) > 0:
    plt.plot(np.real(z2), np.imag(z2), 'o', markersize=12, markeredgewidth=2.5,
             fillstyle='none', label='Ceros', color='red')

plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
circle2 = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--', linewidth=1, alpha=0.5)
plt.gca().add_patch(circle2)
plt.title('Diagrama de Polos y Ceros', fontsize=12, fontweight='bold')
plt.xlabel('Parte Real σ [1/s]', fontsize=11)
plt.ylabel('Parte Imaginaria jω [rad/s]', fontsize=11)
plt.grid(True, alpha=0.7)
plt.legend(fontsize=10, loc='best')
plt.axis('equal')

plt.suptitle('T₂(s)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

#%% T3

num3 = [1, 1/5, 1]
den3 = [1, np.sqrt(2), 1]

# Respuesta en frecuencia
w3, h3 = signal.freqs(num3, den3, worN=np.logspace(-1, 2, 1000))

# Cálculo de fase
phase3 = np.unwrap(np.angle(h3))

# Polos y ceros
z3, p3, k3 = signal.tf2zpk(num3, den3)

# Crear figura
plt.figure(figsize=(14, 5))

# Magnitud
plt.subplot(1, 3, 1)
plt.semilogx(w3, 20*np.log10(abs(h3)), linewidth=2)
plt.title('Respuesta en Magnitud', fontsize=12, fontweight='bold')
plt.xlabel('Frecuencia angular ω [rad/s]', fontsize=11)
plt.ylabel('|H(jω)| [dB]', fontsize=11)
plt.grid(True, which='both', ls=':', alpha=0.7)

# Fase
plt.subplot(1, 3, 2)
plt.plot(w3, phase3/np.pi, linewidth=2)
plt.title('Respuesta en Fase', fontsize=12, fontweight='bold')
plt.xlabel('Frecuencia angular ω [rad/s]', fontsize=11)
plt.ylabel('∠H(jω) [rad]', fontsize=11)
plt.xlim([0, 10])
plt.grid(True, which='both', ls=':', alpha=0.7)
# Configurar formato de ticks en múltiplos de π
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_func))

# Diagrama de polos y ceros
plt.subplot(1, 3, 3)
plt.plot(np.real(p3), np.imag(p3), 'x', markersize=12, markeredgewidth=2.5,
         label='Polos', color='blue')
if len(z3) > 0:
    plt.plot(np.real(z3), np.imag(z3), 'o', markersize=12, markeredgewidth=2.5,
             fillstyle='none', label='Ceros', color='red')

plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
circle3 = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--', linewidth=1, alpha=0.5)
plt.gca().add_patch(circle3)
plt.title('Diagrama de Polos y Ceros', fontsize=12, fontweight='bold')
plt.xlabel('Parte Real σ [1/s]', fontsize=11)
plt.ylabel('Parte Imaginaria jω [rad/s]', fontsize=11)
plt.grid(True, alpha=0.7)
plt.legend(fontsize=10, loc='best')
plt.axis('equal')

plt.suptitle('T₃(s)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
